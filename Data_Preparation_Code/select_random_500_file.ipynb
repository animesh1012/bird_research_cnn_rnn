## Chunking and saving

```python
import os
import librosa
import soundfile as sf

def create_audio_chunks(input_dir, list_dir, output_dir, chunk_duration=15, max_chunks=500, sample_rate=22050):
    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Initialize counters
    total_chunks_created = 0
    
    
    # Traverse through each class directory in list_dir
    for dir_name in list_dir:
        # Full path to the class directory
        class_dir = os.path.join(input_dir, dir_name)
        
        # Create a subdirectory for this class in the output directory
        class_output_dir = os.path.join(output_dir, dir_name)
        os.makedirs(class_output_dir, exist_ok=True)
        
        total_chunks_created = 0
        file_counter = 0
        # Process each file in the class directory
        for file in os.listdir(class_dir):
            # Check if file is an mp3 and we haven't reached max file or chunk count
            if file.endswith(".mp3") and total_chunks_created < max_chunks:
                try:
                    # Construct full file path
                    file_path = os.path.join(class_dir, file)
                    
                    # Load audio file with librosa
                    audio, sr = librosa.load(file_path, sr=sample_rate)
                    
                    # Calculate the number of samples per chunk
                    samples_per_chunk = chunk_duration * sr
                    num_chunks = len(audio) // samples_per_chunk

                    # Create chunks and save them
                    for i in range(num_chunks):
                        if total_chunks_created >= max_chunks:
                            break  # Stop if max chunks limit is reached
                        
                        start_sample = i * samples_per_chunk
                        end_sample = start_sample + samples_per_chunk
                        chunk = audio[start_sample:end_sample]
                        
                        # Define output path for each chunk in the class-specific directory
                        chunk_filename = f"{os.path.splitext(file)[0]}_chunk{i+1}.mp3"
                        chunk_output_path = os.path.join(class_output_dir, chunk_filename)
                        
                        # Save the chunk
                        sf.write(chunk_output_path, chunk, sr)
                        
                        # Increment the chunk counter
                        total_chunks_created += 1

                    # Increment the file counter
                    file_counter += 1
                    print(f"Processed and chunked: {file_path} into {min(num_chunks, max_chunks - total_chunks_created)} chunks.")
                    
                except Exception as e:
                    print(f"Error processing {file}: {e}")

    print(f"Total {file_counter} files processed and {total_chunks_created} chunks saved to {output_dir}.")

# Usage
input_directory = "../data/xeno-canto-dataset-full-all-Countries"
output_directory = "./xeno-canto-clipped_15s"
# class_name = ["class1", "class2", "class3"]  # Replace with actual class names or get dynamically
create_audio_chunks(input_directory, class_name, output_directory)
```

```python
import os

def limit_files_in_directories(base_dir, max_files_per_dir=500):
    # Ensure base_dir exists
    if not os.path.isdir(base_dir):
        raise ValueError(f"The directory {base_dir} does not exist.")
    
    # List all directories in the base_dir
    all_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]
    
    # Iterate over directories
    for dirname in all_dirs:
        dir_path = os.path.join(base_dir, dirname)
        try:
            files = os.listdir(dir_path)
        except Exception as e:
            print(f"Error reading directory {dir_path}: {e}")
            continue
        
        if len(files) > max_files_per_dir:
            excess_files = files[max_files_per_dir:]
            
            # Delete excess files
            for file in excess_files:
                file_path = os.path.join(dir_path, file)
                try:
                    os.remove(file_path)
                except Exception as e:
                    print(f"Error removing file {file}: {e}")
            
            print(f"Removed {len(excess_files)} excess files from {dirname}")

    print("File limiting complete.")



# Example usage
base_directory = './xeno-canto-clipped_15s'
limit_files_in_directories(base_directory)
```
